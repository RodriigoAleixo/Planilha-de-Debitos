```python
time.sleep(10)

#Importanto a planilha.
workbook = openpyxl.load_workbook('cotas_inicio_proc.xlsx')
#Indicando qual aba eu quero utilizar.
pagina_clientes = workbook['Teste']
#Informando qual a primeira linha de leitura
for linha in pagina_clientes.iter_rows(min_row=2):
    grupo = linha [0].value
    cota = linha[1].value
    versao = linha[2].value

    #Calcular Honorários?
    pyautogui.press('tab')
    time.sleep(1)
    pyautogui.press('down')

     #Aguardando carregar:
    while True:
        try:
            carregando = pyautogui.locateOnScreen('carregando.png', grayscale= True, confidence= 0.9) 
            if carregando is None:
                break
        except pyautogui.ImageNotFoundException:
            break
    time.sleep(1)

    #Grupo:
    pyautogui.press('tab', presses=3)
    time.sleep(1)
    pyautogui.write('00')
    pyautogui.write(str(grupo))
    time.sleep(1)
    pyautogui.press('tab')

    #Aguardando carregar:
    while True:
        try:
            carregando = pyautogui.locateOnScreen('carregando.png', grayscale= True, confidence= 0.9) 
            if carregando is None:
                break
        except pyautogui.ImageNotFoundException:
            break
    time.sleep(1)

    #Cota:
    pyautogui.press('tab', presses=4)
    time.sleep(1)
    pyautogui.write(str(cota))
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(1)

    #Aguardando carregar:
    while True:
        try:
            carregando = pyautogui.locateOnScreen('carregando.png', grayscale= True, confidence= 0.9) 
            if carregando is None:
                break
        except pyautogui.ImageNotFoundException:
            break
    time.sleep(1)

    #Versão:
    pyautogui.press('tab', presses=5)
    time.sleep(1)
    pyautogui.write(str(versao))
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(1)

    #Aguardando carregar:
    while True:
        try:
            carregando = pyautogui.locateOnScreen('carregando.png', grayscale= True, confidence= 0.9) 
            if carregando is None:
                break
        except pyautogui.ImageNotFoundException:
            break
    time.sleep(1)

    #Exibir Relatório
    pyautogui.press('tab', presses=9)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)

    #Aguardando carregar:
    while True:
        try:
            carregando = pyautogui.locateOnScreen('carregando.png', grayscale= True, confidence= 0.9) 
            if carregando is None:
                break
        except pyautogui.ImageNotFoundException:
            break
    time.sleep(1)

    #Salvar
    x, y, largura, altura = pyautogui.locateOnScreen('salvar.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(3)

    #PDF
    pyautogui.press('down', presses=3)
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(3)

    #Manter
    x, y, largura, altura = pyautogui.locateOnScreen('manter.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(2)
    pyautogui.press('esc')
    time.sleep(1)
    pyautogui.press('f5')
    time.sleep(2)
```
